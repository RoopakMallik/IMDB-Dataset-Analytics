```python
highest_voted_movie=df['Votes'].max()
df[df['Votes']==highest_voted_movie]
```

```python
highest_revenue_movie=df['Revenue (Millions)']==highest_revenue
df[highest_revenue_movie]
```

```python
highest_runtime=df['Runtime (Minutes)'].max()
highest_runtime
```
